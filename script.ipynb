```python
# both paths must be in the same drive
path_to_captures = r'D:\Xbox\XboxCaptureCopies'
path_save_at = r'D:\Xbox\python'
```

```python
# format name function
def format_name(filename_str):
    # correct date
    old_date = filename_str.split('-')[1]
    new_date = old_date.split('_')[2] + '.' + old_date.split('_')[1] + '.' + old_date.split('_')[0]

    # correct time
    old_time = filename_str.split('-')[2]
    new_time = old_time.split('_')[0] + '.' + old_time.split('_')[1] + '.' + old_time.split('_')[2].split('.')[0]

    # correct title
    old_title = filename_str.split('-')[0]
    new_title = old_title.replace('_', ' ').replace('  ', ' ').replace('™', '').replace('®', '')  #.replace('\'', '')

    # return the new name
    return new_title + ' - ' + new_date + ' - ' + new_time

# rename function
def my_rename(filename_str):
    # main loop
    for filename in os.listdir(filename_str):
        if filename.endswith(".png"):
            # create new name
            new_name = format_name(filename)
            
            # create new path if not exists
            try: 
                os.mkdir(path_save_at + '\png\\' + new_name.split('-')[0].strip())
            except FileExistsError:
                print('Folder already exists:', new_name.split('-')[0].strip())

            # rename images
            try: 
                os.rename(path_to_captures + '\\' + filename, path_save_at + '\png\\' + new_name.split('-')[0].strip() + '\\' + new_name + '.png')
            except FileExistsError:
                print('File already exists')
        
        elif filename.endswith(".mp4"):
            # create new name
            new_name = format_name(filename)
            
            # create new path if not exists
            try: 
                os.mkdir(path_save_at + '\mp4\\' + new_name.split('-')[0].strip())
            except FileExistsError:
                print('Folder already exists:', new_name.split('-')[0].strip())

            # rename images
            try: 
                os.rename(path_to_captures + '\\' + filename, path_save_at + '\mp4\\' + new_name.split('-')[0].strip() + '\\' + new_name + '.mp4')
            # except OSError:
                # print('check the error')
            except FileExistsError:
                print('File already exists')
        else: 
            print('not a png or mp4')
```

```python
# check how many items are in the folder
size = len(os.listdir(path_to_captures))
print('Number of files in the folder: ' + str(size))

# start timer
start_time = time.time()

if size == 0:
    print('Folder is empty')
else: 
    print('Starting to rename files...')
    print('')
    my_rename(path_to_captures)

print('Done')
print('Time elapsed: ' + str(round(time.time() - start_time, 3)) + ' seconds')
```
